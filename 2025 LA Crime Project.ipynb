# Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
crime = pd.read_csv('/Users/bobbiewilliams/Desktop/SQL And Phyton 2025 projects/Data for Phyton Course/Updated Crime_Data_from_2020_to_Present_20250607.csv')
#source https://data.lacity.org/Public-Safety/Crime-Data-from-2020-to-Present/2nrs-mtv8/about_data
```

# Understand and Clean Data

```python
crime.head()
```

```python
crime.shape #<-- Identifying how many rows and columns
```

```python
#Checking Nulls
crime.isna().sum()
```

```python
#Creating a new data frame getting rid of columns with nulls
df = crime.drop(labels = ['Mocodes','Premis Cd','Premis Desc',
                          'Weapon Used Cd','Weapon Desc','Status','Crm Cd 1',
                          'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4','Cross Street'], axis = 1)
df
```

```python
df.sort_values(by = 'Date Rptd')
```

```python
df.dtypes
```

```python
#Update Datatypes. Covert from object to date format
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
```

```python
df.info()
```

```python
df.describe(include = ['float64','int64'])
```

```python
#counting to see the victims age and removing anything below '0'
df['Vict Age'].value_counts()
```

```python
df_age = df[df['Vict Age'] > 0]
```

```python
df_age['Vict Age'].value_counts()
```

```python
#understanding how many years & months and adding it to the data frame. What year the crime occured and month 
df['year'] = df['DATE OCC'].dt.year
df['month'] = df['DATE OCC'].dt.month
```
